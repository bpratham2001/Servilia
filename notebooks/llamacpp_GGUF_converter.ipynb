```python
from llama_cpp import Llama

llm = Llama(model_path="Tulu-3.1-8B.gguf")
response = llm.create_chat_completion(
  messages=[
    {
        "role": "user",
        "content": "Hello, do you know what day it is today?"
    }
])
print(response)
```
